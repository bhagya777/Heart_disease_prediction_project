### Heart Disease Frequency according to Sex

```python
# 1 is for male, 0 for female.
df.sex.value_counts()
```

```python
pd.crosstab(df.sex,df.target)
```

```python
pd.crosstab(df.target,df.sex).plot(kind="bar",color=["salmon","lightblue"])
plt.title("Heart Disease Frequency according to sex")
plt.legend(["Female","male"])
plt.xlabel("Target (0=No Heart Disease, 1=Heart Disease)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

### Age vs Max Heart Rate for Heart Disease

```python
plt.figure(figsize=(10,6))
plt.scatter(df.age[df.target==1],
            df.thalach[df.target==1],
            c="lightblue")
plt.scatter(df.age[df.target==0],
           df.thalach[df.target==0],
           c="salmon")
plt.title("Age va Max Heart Rate for Heart Disease")
plt.legend(["Heart Disease","No Heart Disease"])
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
# Distribution of age 
df.age.plot.hist()
plt.show()
```

```python
### Heart Disease Frequency According to Chest Pain
pd.crosstab(df.cp,df.target)
```

```python
pd.crosstab(df.cp,df.target).plot(kind="bar",color=["salmon","lightblue"],figsize=(10,6))
plt.title("Heart Disease Frequency According to Chest Pain")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(["No Heart Disease","Heart Disease"])
plt.xticks(rotation=0)
plt.show()
```
